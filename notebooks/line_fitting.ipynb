```python
import numpy as np

import matplotlib.pyplot as plt
```

```python
x = np.linspace(0, 10, 11)

y1 = x
y2 = 2 * x
y3 = -2 * x
y4 = x - 10
plt.scatter(x, y1, label='y1')
plt.scatter(x, y2, label='y2')
plt.scatter(x, y3, label='y3')
plt.scatter(x, y4, label='y4')


plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.show()
```
