```python
import numpy as np
import json

from matplotlib import pyplot as plt

np.random.seed(78)
```

```python
noise = np.random.normal(0, 1, 101)
x = np.linspace(0, 10, 101)
y = 1.5 * x + 1 + noise

plt.scatter(x, y, s=5)
plt.xlabel('x')
plt.ylabel('y')
```

```python
# starting parameters
a = 0.3
b = 4.0

plt.plot(x, a * x + b, color='magenta')
plt.scatter(x, y, s=5)
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
# loss function (SSE)
def compute_loss(y_true, y_pred):
    return np.sum(np.square(y_true - y_pred))

y_pred = a * x + b
compute_loss(y, y_pred)
```

```python
NEPOCH = 100
LEARNING_RATE = 0.0001

loss_history = []
for i in range(NEPOCH):
    y_pred = a * x + b

    a = a - LEARNING_RATE * np.sum((y_pred - y) * x)
    b = b - LEARNING_RATE * np.sum(y_pred - y)

    current_loss = compute_loss(y, y_pred)
    loss_history.append(current_loss)
```

```python
# final parameters
plt.plot(x, a * x + b, color='magenta')
plt.scatter(x, y, s=5)
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
plt.plot(loss_history)
plt.yscale('log')
plt.show()
```

```python
# save loss history
with open("results/linear_regression.numpy.result.json", "w") as file_handle:
    json.dump({"a": a, "b": b, "loss_history": loss_history}, file_handle)
```
